```python
import pandas as pd
df = pd.DataFrame({"gruplar1": ["A","B","C","A","B","C"],
                  "degisken1": [10,23,33,22,11,99],
                  "degisken2": [100,253,333,262,11,969]},
                 columns = ["gruplar1","degisken1","degisken2"])
df
```

```python
def filter_func(x) :
    return x["degisken1"].std() >9
    
```

```python
df.groupby("gruplar1").std()
```

```python
df.groupby("gruplar1").filter(filter_func)
```
